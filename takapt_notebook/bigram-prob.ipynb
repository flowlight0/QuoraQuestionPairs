```python
import pandas as pd
import numpy as np
import nltk
from collections import Counter
from nltk.corpus import stopwords
from sklearn.metrics import log_loss
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from scipy.optimize import minimize
stops = set(stopwords.words("english"))
import xgboost as xgb
from sklearn.cross_validation import train_test_split
import multiprocessing
import difflib
```

```python
features = [
#     'clean_q1_tokenized',
#     'clean_q2_tokenized',
#     'clean_q1_stem',
#     'clean_q2_stem',
#     'clean_q1_pos_tagged',
#     'clean_q2_pos_tagged',
    'clean_q1_lemmatized',
    'clean_q2_lemmatized',
#     'clean_q1_lemmatized_stem',
#     'clean_q2_lemmatized_stem'
]
```

```python
def read_data(t, features):
    data = pd.read_csv('../input/{}.csv'.format(t))
    for feature in features:
        data = pd.merge(data, pd.read_pickle('../feature/{}_{}.pkl'.format(t, feature)))
    data.fillna(0.0)
    return data
```

```python
train = read_data('train', features)#[:1000]
test = read_data('test', features)#[:1000]
```

```python
from nltk import FreqDist
bigrams = []
for index, row in train.iterrows():
    bigrams.extend(nltk.bigrams(row['clean_q1_lemmatized']))
    bigrams.extend(nltk.bigrams(row['clean_q2_lemmatized']))
bigram_freq = FreqDist(bigrams)
uniq_bigrams = set(bigrams)
```

```python
bigram_freq.most_common()[int(len(bigram_freq) * 0.2)]
```

```python
def get_weight(bigram):
    return 1 - bigram_freq[bigram] / len(uniq_bigrams)
```

```python
def sum_weight_common_bigrams(row):
    bigrams1 = set(nltk.bigrams(row.clean_q1_lemmatized))
    bigrams2 = set(nltk.bigrams(row.clean_q2_lemmatized))
    common_bigrams = (bigrams1 & bigrams2)
    if common_bigrams:
        return sum(get_weight(bigram) for bigram in common_bigrams)
    else:
        return 0
```

```python
def sum_weight_common_bigrams_limit(row):
    bigrams1 = set(nltk.bigrams(row.clean_q1_lemmatized))
    bigrams2 = set(nltk.bigrams(row.clean_q2_lemmatized))
    common_bigrams = (bigrams1 & bigrams2)
    if common_bigrams:
        return sum(get_weight(bigram) for bigram in common_bigrams if bigram_freq[bigram] >= 2)
    else:
        return 0
```

```python
train['sum_weight_common_bigrams'] = train.apply(sum_weight_common_bigrams, axis=1)
test['sum_weight_common_bigrams'] = test.apply(sum_weight_common_bigrams, axis=1)
```

```python
train['sum_weight_common_bigrams_limit3'] = train.apply(sum_weight_common_bigrams_limit, axis=1)
test['sum_weight_common_bigrams_limit3'] = test.apply(sum_weight_common_bigrams_limit, axis=1)
```

```python
import pandas as pd
import numpy as np
import nltk
from collections import Counter
from nltk.corpus import stopwords
from sklearn.metrics import log_loss
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from scipy.optimize import minimize
stops = set(stopwords.words("english"))
import xgboost as xgb
from sklearn.cross_validation import train_test_split
import multiprocessing
import difflib
```

```python
features = [
#     'clean_q1_tokenized',
#     'clean_q2_tokenized',
#     'clean_q1_stem',
#     'clean_q2_stem',
#     'clean_q1_pos_tagged',
#     'clean_q2_pos_tagged',
    'clean_q1_lemmatized',
    'clean_q2_lemmatized',
#     'clean_q1_lemmatized_stem',
#     'clean_q2_lemmatized_stem'
]
```

```python
def read_data(t, features):
    data = pd.read_csv('../input/{}.csv'.format(t))
    for feature in features:
        data = pd.merge(data, pd.read_pickle('../feature/{}_{}.pkl'.format(t, feature)))
    data.fillna(0.0)
    return data
```

```python
train = read_data('train', features)#[:10000]
test = read_data('test', features)#[:10000]
```

```python
# from nltk import FreqDist
# bigram = []
# for index, row in train.iterrows():
#     bigram.extend(nltk.bigrams(row['clean_q1_lemmatized']))
# bigram_freq = FreqDist(bigram)
def gen_ngrams(words):
    return nltk.ngrams(words, 3)
```

```python
def count_common_ngrams(row):
    a = set(gen_ngrams(row['clean_q1_lemmatized']))
    b = set(gen_ngrams(row['clean_q2_lemmatized']))
    return len(a & b)
```

```python
def jaccard_common_ngrams(row):
    a = set(gen_ngrams(row['clean_q1_lemmatized']))
    b = set(gen_ngrams(row['clean_q2_lemmatized']))
    if not a or not b:
        return 0
    return len(a & b) / len(a | b)
```

```python
def nazo_common_ngrams(row):
    a = set(gen_ngrams(row['clean_q1_lemmatized']))
    b = set(gen_ngrams(row['clean_q2_lemmatized']))
    if not a or not b:
        return 0
    return len(a & b) / (len(a) + len(b))
```

```python
train['common_ngrams_clean_lemm'] = train.apply(count_common_ngrams, axis=1)
test['common_ngrams_clean_lemm'] = test.apply(count_common_ngrams, axis=1)
```

```python
train['jaccard_common_ngrams_clean_lemm'] = train.apply(jaccard_common_ngrams, axis=1)
test['jaccard_common_ngrams_clean_lemm'] = test.apply(jaccard_common_ngrams, axis=1)
```

```python
train['nazo_common_ngrams_clean_lemm'] = train.apply(nazo_common_ngrams, axis=1)
test['nazo_common_ngrams_clean_lemm'] = test.apply(nazo_common_ngrams, axis=1)
```

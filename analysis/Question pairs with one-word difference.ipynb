```python
import pandas as pd
```

```python
df_train = pd.read_csv('../data/input/train.csv.tokenized')
df_test = pd.read_csv('../data/input/test.csv.tokenized')
```

```python
from tqdm import tqdm
one_word_different_pairs = []
for i , (q1, q2, label) in tqdm(enumerate(zip(df_train.question1.astype(str), df_train.question2.astype(str), df_train.is_duplicate))):
    set_q1 = set(q1.split(" "))
    set_q2 = set(q2.split(" "))
    diff_q1 = set_q1.difference(set_q2)
    diff_q2 = set_q2.difference(set_q1)
    if len(diff_q1) <= 1 and len(diff_q2) <= 1:
        one_word_different_pairs.append([i, q1, q2, diff_q1, diff_q2, label])
```

```python
len(one_word_different_pairs)
```

```python
df = pd.DataFrame(data=one_word_different_pairs, columns=['train_id','q1', 'q2', 'diff_q1', 'diff_q2', 'is_duplicate'])
```

```python
df
```

```python
train_pred = pd.read_csv('../data/output/gbm_70.json.gbm_70.json.submission.csv.train.pred.csv')
train_pred.rename(columns={'id': 'train_id', 'is_duplicate': 'prediction'}, inplace=True)
```

```python
df = df.merge(train_pred, on='train_id', how='inner')
```
